```python
from tqdm import tqdm
import time
from sympy import primerange

def ModRoot(a,n):
    
    r=[]
    amod=a%n
    for m in range(n):
        if (m*m)%n==amod:
            r.append(m)
            
    return r
    
Nmax=50
Nmin=5
prs=list(primerange(1,24))
print(prs)

for n in tqdm(range(Nmin,Nmax+1)):
    for a in prs:
        for b in prs:
            if b>=a:
                for c in prs:
                    if c!=a and c!=b:
                        Moda=ModRoot(a,n)
                        if Moda!=[]:
                            Modb=ModRoot(b,n)
                            if Modb!=[]:
                                Modc=ModRoot(c,n)
                                if Modc!=[]:
                                    for ar in Moda:
                                        for br in Modb:
                                            for cr in Modc:
                                                if ar!=0 and br!=0:
                                                    if cr!=0:
                                                        if (ar+br)%n==cr:
                                                            print(n)
                                                            print("root:",a,b,c)
                                                            print("ModRoot",ar,br,cr)
```
